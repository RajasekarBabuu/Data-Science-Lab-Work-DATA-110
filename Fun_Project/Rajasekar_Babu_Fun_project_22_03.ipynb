```python
#imports
import qrcode
from PIL import Image
from IPython.display import display

print(" Peace QR Generator ")

# Inputs
name = input("Enter your name: ")
country = input("Enter your country: ")
comments = input("Enter your comments: ")

# Your message (English)
message = """This is Rajasekar Babu.

WE want peace.
I have created this as a small new initiative.
Please take a moment to look into it.

Thank you."""

# Show in console
print("\n--- Peace Message ---")
print(message)

#  QR will open BOTH message + link
data = message + "\n\n" + "https://kingsek.github.io/peace-project/"

# Generate QR
qr = qrcode.make(data)

filename = name + "_peace_qr.png"
qr.save(filename)

print(" QR Code created!")

# Show QR
img = Image.open(filename)
display(img)
```
